```python
class Property:
  def __init__(self, p_type, p_city):
    self.p_type = p_type
    self.p_city = p_city
    self.totalInvestment = self.TotalInvestment()
    self.monthlyIncome = self.MonthlyIncome()
    self.monthlyExpenses = self.MonthlyExpenses()

  class TotalInvestment:
    def getTotalInvestment(self):
      totalInvestment = 0
      while True:
        investmentType = input('Please Input Type of Investment: ')
        investmentAmount = int(input(f'Please Input amount for Investment of {investmentType} $:'))
        totalInvestment = totalInvestment + investmentAmount
        response = input('Do you want to add more Investment (y/n)?')
        if (response == 'n' or response == 'N'):
          break
      return totalInvestment

  class MonthlyIncome:
    def getMonthlyIncome(self):
      totalMonthlyIncome = 0
      while True:
        incomeType = input('Please Input Type of Monthly Income: ')
        incomeAmount = int(input(f'Please Input amount for Monthly Income of {incomeType} $:'))
        totalMonthlyIncome = totalMonthlyIncome + incomeAmount
        response = input('Do you want to add more Monthly Income (y/n)?')
        if (response == 'n' or response == 'N'):
          break
      return totalMonthlyIncome

  class MonthlyExpenses:
    def getMonthlyExpenses(self):
      totalMonthlyExpenses = 0
      while True:
        expensesType = input('Please Input Type of Monthly Expenses: ')
        expenseAmount = int(input(f'Please Input amount for Monthly Expenses of {expensesType} $:'))
        totalMonthlyExpenses = totalMonthlyExpenses + expenseAmount
        response = input('Do you want to add more Monthly Expenses (y/n)?')
        if (response == 'n' or response == 'N'):
          break
      return totalMonthlyExpenses
  
  def monthlyCashFlow(self, Montly_Income, Monthly_Expenses):
    return (Montly_Income - Monthly_Expenses)
  
  def ROI(self):
    return ((12*self.monthlyCashFlow(self.monthlyIncome,self.monthlyExpenses)*100)/self.totalInvestment)

  def printInvestmentAnalysis(self):
    print(f'\n****** Analysis of your Rental income form Proposed Investment in {self.p_type} at {self.p_city} is as below ******')
    print('**Monthly Income**')
    print(f'Total Monthly Income = ${self.monthlyIncome}')
    print('**Monthly Expenses**')
    print(f'Total Monthly Expenses = ${self.monthlyExpenses}')
    print('**Monthly Cash Flow**')
    print(f'Total Monthly Cash Flow = ${self.monthlyCashFlow(self.monthlyIncome,self.monthlyExpenses)}')
    print('**Total Investment**')
    print(f'Total Investment Amount = ${self.totalInvestment}')
    print('\n**Return On Investment**')
    print(f'ROI = {self.ROI():.2f}%') 

  def getData(self):
    print("\nPlease input Monthly Income Details")
    self.monthlyIncome = self.monthlyIncome.getMonthlyIncome()
    print("\nPlease input Monthly Expenses Details")
    self.monthlyExpenses = self.monthlyExpenses.getMonthlyExpenses()
    print("\nPlease input Total Investment Details")
    self.totalInvestment = self.totalInvestment.getTotalInvestment()

try:
  print("Please input Property Details")
  propertyType = input('Please input type of Property: ')
  propertyCity = input('Please input Property City: ')

  newProperty = Property(propertyType, propertyCity)

  newProperty.getData()

  newProperty.printInvestmentAnalysis()
except:
  print("Something went wrong, Please try again and input correct infromation")
```
